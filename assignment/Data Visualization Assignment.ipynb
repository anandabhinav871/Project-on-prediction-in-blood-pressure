## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
ind=[x for x, _ in enumerate(store)]
plt.bar(ind,Clothing,width=0.8,label='Clothing',color='darkorange',bottom=Equipment+Accessories)
plt.bar(ind,Equipment,width=0.8,label='Equipments',color='lightblue',bottom=Accessories)
plt.bar(ind,Accessories,width=0.8,label='Accessories',color='red')
plt.xticks(ind,store,rotation=90)
plt.xlabel("Cities")
plt.ylabel("Stores")
plt.legend(loc='upper right')
plt.title("Stores in various cities")
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
d=sns.load_dataset('titanic')
d.head()
```

```python
s=sns.countplot(x='pclass',hue='survived',palette='Set1',data=d)
s.set(title='Survival of passenger against the Passenger class')
plt.legend(loc="upper left")
plt.show()
```

## Q.6 For Tips Dataset

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
tip=sns.load_dataset('tips')
tip.head()
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=1)
plt.title('Days Vs Total Bill')
plt.xticks(rotation=45)
```
